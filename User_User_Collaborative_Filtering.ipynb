```python
import requests

# The Google Sheets document ID from the URL
file_id = '1Moo1GrIu2K24kRZSS1QyLAkUXtUwiS3N'

# The URL for downloading the Excel file
url = f'https://docs.google.com/spreadsheets/d/{file_id}/export?format=xlsx'

# Make a request to the URL
response = requests.get(url)

# Save the content as an Excel file
with open('spreadsheet_data.xlsx', 'wb') as file:
    file.write(response.content)
```

```python
import pandas as pd
data_sheet = pd.read_excel('spreadsheet_data.xlsx')
data_sheet.head()
```

```python
means_dict = {}
for column in data_sheet.columns[1:]:
    means_dict[column] = data_sheet[column].mean()
```
